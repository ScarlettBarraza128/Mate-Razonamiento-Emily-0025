#### Problema 1 Paso 1: Análisis y resolución algebraicaPara resolver esto, primero debemos "traducir" el lenguaje común al lenguaje matemático.Planteamiento: Definimos nuestra incógnita. Sea $x$ el número que buscamos.Ecuación: La frase "la suma de un número más 9 es igual a 21" se escribe como:$$x + 9 = 21$$Despeje: Para dejar sola a la $x$, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Solución:$$x = 12$$

```python
# Problema 1: Hallar un número x tal que x + 9 = 21
# resultado_total = 21
# numero_conocido = 9
resultado_total=int(input("Ingrese el resultado total: "))
numero_conocido=int(input("Ingrese el numero conocido: "))
# Despejamos: x = 21 - 9
x = resultado_total - numero_conocido

print(f"El número buscado es: {x}")
```

#### Problema 2
Planteamiento:Edad de Ana: $x$Edad de Pedro: $2x$ (porque es el doble que la de Ana)Ecuación: "Entre los dos suman un total de 36 años":$$x + 2x = 36$$Despeje:Simplificamos las $x$: $3x = 36$El 3 que multiplica a la $x$ pasa dividiendo:$$x = \frac{36}{3}$$Solución:$x = 12$ (Edad de Ana)$2x = 2(12) = 24$ (Edad de Pedro)

```python
# Problema 2: Ana (x) + Pedro (2x) = 36
# suma_edades = 36
suma_total=int(input("Ingrese la suma total: "))
# Despejamos: x + 2x = 36

# La ecuación simplificada es 3x = 36, por lo tanto x = 36 / 3
edad_ana = suma_total / 3
edad_pedro = edad_ana * 2

print(f"Ana tiene {int(edad_ana)} años")
print(f"Pedro tiene {int(edad_pedro)} años")
```
